```python
import pandas as pd
import numpy as np
import random


# Leer tu CSV
df = pd.read_csv('la hora del amor.csv')
```

```python
df.head()
```

```python
# Crear una columna aleatoria llamada "Género"
df['Género'] = np.random.choice(
    ['Hombre', 'Mujer', 'No binario'],
    size=len(df),
    p=[0.55, 0.35, 0.10]  # 45%, 45%, 10%
)
```

```python
df['age']= df['age'].astype(int)
```

```python
#borrar columa errónea

df = df.drop(columns=['gender'])
```

```python
#Provincias Totalmente Random
provincias_espana = [
    "Álava", "Albacete", "Alicante", "Almería", "Asturias", "Ávila",
    "Badajoz", "Barcelona", "Burgos", "Cáceres", "Cádiz", "Cantabria",
    "Castellón", "Ciudad Real", "Córdoba", "Cuenca", "Girona", "Granada",
    "Guadalajara", "Gipuzkoa", "Huelva", "Huesca", "Illes Balears", "Jaén",
    "La Coruña", "La Rioja", "Las Palmas", "León", "Lleida", "Lugo",
    "Madrid", "Málaga", "Murcia", "Navarra", "Ourense", "Palencia",
    "Pontevedra", "Salamanca", "Santa Cruz de Tenerife", "Segovia",
    "Sevilla", "Soria", "Tarragona", "Teruel", "Toledo", "Valencia",
    "Valladolid", "Bizkaia", "Zamora", "Zaragoza"
]

df['Provincia'] = [random.choice(provincias_espana) for _ in range(len(df))]
```

```python
# Guardar el CSV nuevo
df.to_csv('hora_amor.csv', index=False)

print(df.head())
```
